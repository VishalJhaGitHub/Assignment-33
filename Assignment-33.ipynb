```python
#1. What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

#Ans

#PMF

#The PMF is a function that describes the probability of a discrete random variable taking on a specific value. It maps each possible outcome to its probability. The PMF is defined for discrete random variables because it assigns probabilities to individual outcomes, rather than to ranges of outcomes. The sum of all probabilities assigned by the PMF must equal 1.

#For example, suppose we toss a fair six-sided die. The PMF of the random variable X, which represents the number rolled, is:

#PMF(X = 1) = 1/6
#PMF(X = 2) = 1/6
#PMF(X = 3) = 1/6
#PMF(X = 4) = 1/6
#PMF(X = 5) = 1/6
#PMF(X = 6) = 1/6

#This means that the probability of rolling a 1 is 1/6, the probability of rolling a 2 is 1/6, and so on.

#PDF

#The PDF is a function that describes the probability density of a continuous random variable at a given point. Unlike the PMF, the PDF is defined for continuous random variables because it assigns probabilities to ranges of outcomes, rather than to individual outcomes. The area under the PDF curve over a given range of values represents the probability of the random variable taking on values within that range.

#For example, suppose we have a random variable Y that is uniformly distributed between 0 and 1. The PDF of Y is:

#PDF(Y) = 1 for 0 <= Y <= 1
#0 otherwise

#This means that the probability density of Y is constant over the range [0,1], and zero outside that range. The area under the PDF curve between any two points a and b in [0,1] is simply the length of the interval [a,b], because the PDF is constant over that interval. For example, the probability that Y takes on a value between 0.2 and 0.6 is:

#P(0.2 <= Y <= 0.6) = ∫0.6_0.2 PDF(Y) dY
#= ∫0.6_0.2 1 dY
#= 0.6 - 0.2
#= 0.4
```

```python
#2. What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

#Ans

#The Cumulative Density Function (CDF) is a function that describes the cumulative probability of a random variable taking on a value less than or equal to a given value. It is defined for both discrete and continuous random variables, and is a fundamental concept in probability theory and statistics.

#For a discrete random variable X, the CDF is defined as:

#F(x) = P(X <= x)

#That is, the CDF gives the probability that X takes on a value less than or equal to x. The CDF is a non-decreasing function, meaning that as x increases, the probability of X being less than or equal to x also increases. The CDF ranges between 0 and 1, with F(-∞) = 0 and F(∞) = 1.

#For a continuous random variable X, the CDF is defined as:

#F(x) = ∫_-∞^x f(t) dt

#where f(t) is the probability density function (PDF) of X. The CDF gives the probability that X takes on a value less than or equal to x. The CDF is also a non-decreasing function, and ranges between 0 and 1, with F(-∞) = 0 and F(∞) = 1.

#Here is an example to illustrate the concept of CDF:

#Suppose we have a continuous random variable X that is uniformly distributed between 0 and 1. The PDF of X is:

#f(x) = 1 for 0 <= x <= 1
#0 otherwise

#The CDF of X is:

#F(x) = ∫_0^x f(t) dt = x for 0 <= x <= 1
#0 otherwise

#This means that the probability that X takes on a value less than or equal to 0.5 is:

#F(0.5) = 0.5

#Similarly, the probability that X takes on a value less than or equal to 0.8 is:

#F(0.8) = 0.8

#The CDF is used in many statistical applications, such as hypothesis testing, confidence intervals, and probability distributions. One reason for its usefulness is that it allows us to calculate probabilities for ranges of values, rather than just for individual values. For example, the probability that X takes on a value between a and b is:

#P(a <= X <= b) = F(b) - F(a)

#Another important property of the CDF is that it uniquely determines the probability distribution of a random variable. In other words, if we know the CDF of a random variable, we can determine its PDF and PMF (if it exists).
```

```python
#3. What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

#Ans

#Some examples of situations where the normal distribution might be used as a model:

#1 - Height and weight: The distribution of human height and weight often follows a normal distribution.

#2 - IQ scores: The distribution of IQ scores is often modeled using a normal distribution.

#3 - Measurement errors: Many types of measurement errors, such as errors in instrumentation or errors due to human factors, can be modeled using a normal distribution.

#4 - Financial data: Many financial data, such as stock prices or currency exchange rates, can be modeled using a normal distribution.

#5 - Physical phenomena: Many physical phenomena, such as the motion of gas molecules, the speed of particles in a fluid, or the intensity of light, can be modeled using a normal distribution.

#The normal distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ). The mean determines the center of the distribution, while the standard deviation determines the spread of the distribution. Specifically, the mean represents the location of the peak of the distribution, while the standard deviation represents the width of the distribution.

#The shape of the normal distribution is symmetric and bell-shaped, with the highest density of data around the mean. The tails of the distribution extend infinitely in both directions, although the probability of extreme values decreases rapidly as we move away from the mean.

#The mean and standard deviation of the normal distribution can be used to determine the probability of a given value occurring. For example, the probability of a random variable X being less than or equal to a certain value x can be calculated using the standard normal distribution table or the z-score formula:

#P(X <= x) = Φ((x - μ) / σ)

#where Φ is the cumulative distribution function of the standard normal distribution. By knowing the mean and standard deviation of the normal distribution, we can make predictions about the likelihood of certain outcomes or estimate the confidence intervals of our measurements.
```

```python
#5. What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

#Ans

#The Bernoulli distribution is a probability distribution that describes the outcomes of a single binary experiment. In other words, it is a distribution that describes the probability of a single event with two possible outcomes, typically denoted as success and failure, or 1 and 0.

#The Bernoulli distribution is named after Swiss mathematician Jacob Bernoulli, who used it to study the probability of coin tosses. The Bernoulli distribution is parameterized by a single parameter p, which represents the probability of success in the experiment.

#The probability mass function (PMF) of the Bernoulli distribution is given by:

#P(X = 1) = p
#P(X = 0) = 1 - p

#where X is a random variable that takes on the value 1 with probability p and the value 0 with probability 1-p.

#An example of a Bernoulli distribution is flipping a fair coin, where success is defined as getting heads and failure is defined as getting tails. In this case, p = 0.5, and the PMF is given by:

#P(X = 1) = 0.5
#P(X = 0) = 0.5

#The Bernoulli distribution is related to the binomial distribution, which is a probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials. The main difference between the two is that the Bernoulli distribution models a single trial, while the binomial distribution models multiple trials.

#The binomial distribution is parameterized by two parameters: the number of trials n and the probability of success in each trial p. The probability mass function (PMF) of the binomial distribution is given by:

#P(X = k) = (n choose k) p^k (1 - p)^(n - k)

#where X is a random variable that takes on the value k, which is the number of successes in n independent Bernoulli trials, (n choose k) is the binomial coefficient, and p is the probability of success in each trial.
```

```python
#8. What is the z score? State the importance of the z score.

#Ans

#The z-score, also known as the standard score, is a statistical measure that represents the number of standard deviations a data point is from the mean of a normal distribution. The z-score is calculated by subtracting the mean from the observed value and dividing the result by the standard deviation.

#The formula for z-score is:

#z = (x - μ) / σ

#where x is the observed value, μ is the mean of the population, and σ is the standard deviation of the population.

#The z-score tells us how many standard deviations an observation is away from the mean of a population. If the z-score is positive, it means that the observation is above the mean, and if the z-score is negative, it means that the observation is below the mean.

#The importance of the z-score is that it allows us to compare data from different populations or datasets that have different means and standard deviations. By standardizing the data using the z-score, we can compare data points from different datasets on a common scale. The z-score also allows us to determine the probability of observing a certain value or range of values in a normal distribution.

#In addition, the z-score is used to identify outliers in a dataset. Any data point with a z-score that is greater than 3 or less than -3 is considered to be an outlier, as it is more than three standard deviations away from the mean.
```

```python
#9. What is Central Limit Theorem? State the significance of the Central Limit Theorem.

#Ans

#The Central Limit Theorem (CLT) is a fundamental result in statistics that states that if we take a large number of random samples from any population, the sampling distribution of the means will be approximately normally distributed, regardless of the shape of the population distribution. Specifically, the CLT states that as the sample size increases, the distribution of the sample means will become increasingly normal in shape, with a mean equal to the population mean and a standard deviation equal to the population standard deviation divided by the square root of the sample size.

#The significance of the Central Limit Theorem is that it allows us to make inferences about a population based on a sample of observations. Since the sampling distribution of the means is approximately normal, we can use the properties of the normal distribution to make statistical inferences about the population, such as calculating confidence intervals and performing hypothesis tests.

#Another important significance of the Central Limit Theorem is that it explains why the normal distribution is so commonly observed in many natural phenomena. Many real-world processes involve a large number of small, random fluctuations, and the CLT shows that the sum of these fluctuations tends to be normally distributed.

#The Central Limit Theorem has wide-ranging applications in many fields, including finance, engineering, and biology. For example, it can be used to estimate the mean and standard deviation of a population from a small sample, or to assess whether a process is within statistical control. It is also used in the development of many statistical methods, such as regression analysis and analysis of variance.
```

```python
#10. State the assumptions of the Central Limit Theorem.

#Ans

#The Central Limit Theorem (CLT) is a powerful statistical tool that allows us to make inferences about a population based on a sample of observations. However, the CLT relies on certain assumptions in order to be valid.

#The assumptions of the Central Limit Theorem are:

#1 - Random Sampling: The sample must be selected at random from the population. Each observation in the sample must be independent of the others and have an equal chance of being selected.

#2 - Independence: The observations in the sample must be independent of each other. This means that the value of one observation does not affect the value of another observation.

#3 - Sample Size: The sample size must be large enough. While there is no strict rule for what constitutes a "large" sample size, a sample size of at least 30 is often used as a guideline.

#4 - Finite Variance: The population from which the sample is drawn must have a finite variance. If the variance of the population is infinite, the CLT may not hold.

#If these assumptions are met, the Central Limit Theorem tells us that the sampling distribution of the means will be approximately normal, regardless of the shape of the population distribution. However, if any of these assumptions are violated, the CLT may not hold and other methods may need to be used to make inferences about the population.
```
